```python
import pandas as pd

# Load the first CSV file
csv_file1 = 'C:/tmp/Exp7/SSIM_train.csv'
df1 = pd.read_csv(csv_file1)

# Load the second CSV file
csv_file2 = 'C:/tmp/Exp7/SSIM_val.csv'
df2 = pd.read_csv(csv_file2)

# Rename the 'Value' columns to distinguish between train and validation
df1.rename(columns={'Value': 'SSIM_train'}, inplace=True)
df2.rename(columns={'Value': 'SSIM_val'}, inplace=True)

# Concatenate the DataFrames
merged_df = pd.merge(df1, df2, on='Step', how='outer')
merged_df.head()


# Load the first CSV file
csv_file1 = 'C:/tmp/Exp7/MSE_train.csv'
df1 = pd.read_csv(csv_file1)

# Load the second CSV file
csv_file2 = 'C:/tmp/Exp7/MSE_val.csv'
df2 = pd.read_csv(csv_file2)

# Rename the 'Value' columns to distinguish between train and validation
df1.rename(columns={'Value': 'MSE_train'}, inplace=True)
df2.rename(columns={'Value': 'MSE_val'}, inplace=True)

# Concatenate the DataFrames
merged_mse = pd.merge(df1, df2, on='Step', how='outer')
merged_df.head()


# Load the first CSV file
csv_file1 = 'C:/tmp/Exp7/Huber_train.csv'
df1 = pd.read_csv(csv_file1)

# Load the second CSV file
csv_file2 = 'C:/tmp/Exp7/Huber_val.csv'
df2 = pd.read_csv(csv_file2)

# Rename the 'Value' columns to distinguish between train and validation
df1.rename(columns={'Value': 'Huber_train'}, inplace=True)
df2.rename(columns={'Value': 'Huber_val'}, inplace=True)

# Concatenate the DataFrames
merged_huber = pd.merge(df1, df2, on='Step', how='outer')
merged_df.head()



```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the merged CSV file


plt.figure(figsize=(10, 5))

# Plot training SSIM
plt.plot(merged_df['Step'], merged_df['SSIM_train'], label='SSIM/Train')

# Plot validation SSIM
plt.plot(merged_df['Step'], merged_df['SSIM_val'], label='SSIM/Val')

# Adding title and labels
plt.title('SSIM Over Training Steps')
plt.xlabel('Step')
plt.ylabel('SSIM')

# Display the legend
plt.legend()

# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# Plot training SSIM
plt.plot(merged_mse['Step'], merged_mse['MSE_train'], label='MSE/Train')

# Plot validation SSIM
plt.plot(merged_mse['Step'], merged_mse['MSE_val'], label='MSE/Val')

# Adding title and labels
plt.title('MSE Over Training Steps')
plt.xlabel('Step')
plt.ylabel('MSE')

# Display the legend
plt.legend()
plt.ylim(0, 0.04)
# Show the plot
plt.show()
```

```python
plt.figure(figsize=(10, 5))

# Plot training SSIM
plt.plot(merged_huber['Step'], merged_huber['Huber_train'], label='MSE/Train')

# Plot validation SSIM
plt.plot(merged_huber['Step'], merged_huber['Huber_val'], label='Huber/Val')

# Adding title and labels
plt.title('Huber Over Training Steps')
plt.xlabel('Step')
plt.ylabel('Huber Loss')

# Display the legend
plt.legend()
plt.ylim(0, 0.04)
# Show the plot
plt.show()
```
